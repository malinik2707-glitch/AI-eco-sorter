```python
import os
import numpy as np
import tensorflow as tf

input_folder = r"C:\Users\ADMIN\Documents\waste_classification\dataset"  
output_folder = r"C:\Users\ADMIN\Documents\resized_image\dataset" 
img_size = (224, 224)  
os.makedirs(output_folder, exist_ok=True)

for category in os.listdir(input_folder):
    category_path = os.path.join(input_folder, category)
    output_category_path = os.path.join(output_folder, category)
    os.makedirs(output_category_path, exist_ok=True)

    for img_name in os.listdir(category_path):
        img_path = os.path.join(category_path, img_name)

        try:
            img = tf.keras.utils.load_img(img_path, target_size=img_size)
        except Exception as e:
            print(f"Skipping {img_name}: {e}")
            continue

        img_array = tf.keras.utils.img_to_array(img) / 255.0

        save_path = os.path.join(output_category_path, img_name)
        tf.keras.utils.save_img(save_path, img_array)

print("All images resized, normalized, and saved!")
```

```python
import os
import shutil
import random

dataset = r"C:\Users\ADMIN\Documents\waste_classification\dataset"
output = r"C:\Users\ADMIN\Documents\splitted_image\dataset"

splits = ['train', 'val', 'test']
ratios = [0.7, 0.15, 0.15]

for split in splits:
    for category in os.listdir(dataset):
        os.makedirs(os.path.join(output, split, category), exist_ok=True)

for category in os.listdir(dataset):
    category_path = os.path.join(dataset, category)
    images = os.listdir(category_path)
    random.shuffle(images)

    total = len(images)
    train_end = int(total * ratios[0])
    val_end = train_end + int(total * ratios[1])

    train_imgs = images[:train_end]
    val_imgs = images[train_end:val_end]
    test_imgs = images[val_end:]

    for img in train_imgs:
        shutil.copy(os.path.join(category_path, img),
                    os.path.join(output, 'train', category, img))

    for img in val_imgs:
        shutil.copy(os.path.join(category_path, img),
                    os.path.join(output, 'val', category, img))

    for img in test_imgs:
        shutil.copy(os.path.join(category_path, img),
                    os.path.join(output, 'test', category, img))
```
